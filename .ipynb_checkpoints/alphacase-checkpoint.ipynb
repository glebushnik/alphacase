```python
from warnings import filterwarnings
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

filterwarnings("ignore")
```

```python
df = pd.read_csv("survey_responses_1000.csv")
df.head()
```

```python
df_melted = pd.melt(df, 
                    id_vars=['name', 'city', 'freq', 'top_func', 'optional_func', 'source', 'new_info_source', 'events', 'gaming'], 
                    value_vars=['Q5_1', 'Q5_2', 'Q5_3', 'Q5_4', 'Q5_5'], 
                    var_name='Q5', 
                    value_name='Importance')

# Убедиться, что все значения важности имеют правильный тип данных (целое число)
df_melted['Importance'] = df_melted['Importance'].astype(int)

csv_path_1000 = "melted_survey_responses_1000.csv"
df_melted.to_csv(csv_path_1000, index=False)

df_melted.drop(columns='Q5')
```

```python
df_melted = df.drop(columns=['name', 'city'])
```

```python

colors = ['#8FFF00', '#188392', '#E7362B', '#FFFF00']
sns.set(style="white")

for column in df_melted.select_dtypes(include=['object']).columns:
    counts = df[column].value_counts()
    plt.figure(figsize=(14, 6))
    
    plt.subplot(1, 2, 1)
    plt.pie(counts, labels=counts.index, colors=colors[:len(counts)], autopct='%1.1f%%', startangle=90)
    plt.title(f'Pie Chart of {column}', fontsize=16)
    
    plt.subplot(1, 2, 2)
    sns.barplot(x=counts.index, y=counts.values, palette=colors[:len(counts)])
    plt.title(f'Bar Chart of {column}', fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Count', fontsize=14)
    
    plt.tight_layout()
    plt.show()
```
